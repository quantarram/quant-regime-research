```python
import pandas as pd

df = pd.read_csv(
    '../data/raw/SPY_daily_yahoo_raw.csv',
    skiprows=[0,1,2], header=None,
    names=['Date','Adj Close','Close','High','Low','Open','Volume'],
    index_col=0, parse_dates=True
)

df = df.sort_index()

# Get 2025 trading days
trading_days_2025 = df.loc['2025-01-01':'2025-12-31'].index

# Pick a start date in 2025
start = trading_days_2025[0]
end = trading_days_2025[199]  # 200th trading day

calendar_days = (end - start).days

print(f"First trading day 2025: {start.date()}")
print(f"200th trading day 2025: {end.date()}")
print(f"Calendar days in between: {calendar_days}")
```
